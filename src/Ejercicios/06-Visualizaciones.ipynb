```python
import pandas as pd
import matplotlib.pyplot as plt   
import seaborn as sns
```

1. Hacer un gráfico que muestre cuántos establecimientos de cada Nivel hay

```python
data = pd.read_csv("establecimientos_educativos.csv")
data["nivel"].value_counts().sort_values(ascending=False)
sns.countplot(data["nivel"])
```

2. Hacer un gráfico que muestre la matrícula de varones por subvencion. Hacer otro gráfico que muestre la matrícula de mujeres por subvencion.

```python
sns.barplot(data=data, x="varones", y="subvencion")
```

```python
sns.barplot(data=data, x="mujeres", y="subvencion")
```

3. Hacer un gráfico para ver la distribución de la matrícula de varones 

```python
sns.distplot(data["varones"])
kind="kde"
```
