### Ejercicios Extra Filtrar un DataFrame con Librería Pandas

Para estos ejercicios vamos a seguir utilizando el recorte de un dataset sobre el establecimientos educativos en la Provincia de Buenos Aires extraído del sitio web de Datos Abiertos de la Provincia de Buenos Aires actualizado a fines del año 2022. 

Deben utilizar el Dataset modificado que usamos en la clase por lo que deben extraerlo del Classroom.

```python
import pandas as pd
data = pd.read_csv("establecimientos_educativos.csv")
data.columns
```

1. Hacer un filtro para ver solamente los establecimientos de dependencia "Privada"

```python
mask_privado = data["sector"] == "Privado"
data_privado = data.loc[mask_privado]

cant_privados = data_privado["row_number"].count()
total = data["row_number"].count()
print("Hay", cant_privados, "establecimientos privados de un total de", total, "establecimientos educativos.")

data_privado
```

2. ¿Cuál es la distrito que menos establecimientos tiene? Hacer un filtro del mismo. ¿Cuántas modalidades tiene? 

```python
estab_min = data["distrito"].min()

mask_alsina = data["distrito"] == "Adolfo Alsina"
data_alsina = data.loc[mask_alsina]

cant_alsina = data_alsina["row_number"].count()
print("El distrito que menos establecimientos tiene es", estab_min, 
        "con un total de", cant_alsina, "establecimientos.")

alsina_modalidades = data_alsina["modalidad"].nunique()
print("\"Adolfo Alsina\" tiene", alsina_modalidades, "modalidades.")
```

3. De la oferta de la Región Educativa I ¿Cuántos establecimientos tiene? ¿Qué modalidades tiene? Hacer un filtro seleccionando solo los establecimientos de modalidad Educación Común

```python
mask_region_1 = data["region_educativa"] == "Región I"
data_region_1 = data.loc[mask_region_1]
cant_region_1 = data_region_1["row_number"].count()
print("La Región I tiene", cant_region_1, "establecimientos.")

modal_region_1 = data_region_1["modalidad"].nunique()
print("La Región I tiene", modal_region_1, "modalidades.")

modal_region_1 = data_region_1["modalidad"] == "Educación Común"
cant_comun_1 = data_region_1.loc[modal_region_1]

print("La Región I tiene los siguientes establecimientos de modalidad \"Educación Común\":\n\n", 
        cant_comun_1)
```

4. Hacer un filtro para seleccionar los establecimientos de nivel "Plan Fines (Trayectos y Deudores), ¿Cuál es la matrícula total de estos establecimientos?

```python
mask_fines = data["nivel"] == "Plan Fines (Trayectos y Deudores)"
data_fines = data.loc[mask_fines]
print("Establecimientos con Plan Fines:\n\n", data_fines)

fines_total = data_fines["matricula"].sum()
print("La matricula total del \"Plan Fines (Trayectos y Deudores)\" es de:", fines_total)
```

5. Hacer un filtro de los establecimientos de Nivel Secundario del area INTERIOR. ¿Cuál es la mátricula de varones y cuál la de mujeres de estos establecimientos?

```python
mask_interior = (data["nivel"] == "Nivel Secundario") & (data["area"] == "INTERIOR")
data_interior = data.loc[mask_interior]
print("Establecimientos de Area Interior:\n\n", data_interior)

interior_varones = data_interior["varones"].sum()
interior_mujeres = data_interior["mujeres"].sum()

print("Matricula de varones:", interior_varones)
print("Matricula de mujeres:", interior_mujeres)

```
