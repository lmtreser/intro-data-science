### Ejercicios Clase 5 - Filtrar Dataset con Pandas

Para estos ejercicios vamos a seguir utilizando el recorte de un dataset sobre el establecimientos educativos en la Provincia de Buenos Aires extraído del sitio web de Datos Abiertos de la Provincia de Buenos Aires actualizado a fines del año 2022. 

Deben utilizar el Dataset modificado que usamos en la clase por lo que deben extraerlo del Classroom.

```python
import pandas as pd
```

```python
data = pd.read_csv("establecimientos_educativos.csv")
pd.options.display.max_columns=None
data
```

1. Hacer un filtro para seleccionar solo los estableciientos de nivel Primario

```python
mask_primario = data["nivel"] == "Nivel Primario"
data_primarios = data.loc[mask_primario]
data_primarios
```

2. Hacer un filtro para seleccionar solo los establecimientos cuyo tipo de organización es "ESCUELA SECUNDARIA" ¿Cuántos establecimientos son? ¿Cuál es la matricula total de estos establecimientos=?

```python
mask_secundario = data["nivel"] == "Nivel Secundario"
data_secundarios = data.loc[mask_secundario]

cantidad_secu = data_secundarios["nombre"].count()
print("Son", cantidad_secu ,"establecimientos de Nivel Secundario.")

matricula_secu = data_secundarios["matricula"].sum()
print("La matricula es de", matricula_secu, "alumnos.")

data_secundarios
```

3. Hacer un filtro para seleccionar solo los establecimientos de Formación Integral. ¿Cuál es la matrícula del establecimiento que más cantidad de estudiantes tiene?

```python
mask_integral = data["nivel"] == "Formación Integral"
data_integrales = data.loc[mask_integral]

matricula_inte = data_integrales["matricula"].max()
print("La matricula del más poblado es de", matricula_inte, "alumnos.")

data_integrales
```

4. Hacer un filtro para seleccionar solo los establecimientos Sin Suvención. ¿Cuál es la matrícula total de estos establecimientos? ¿De varones?

```python
mask_suvencion = data["subvencion"] == "Oficial"
data_suvenciones = data.loc[mask_suvencion]

matricula_suve = data_suvenciones["matricula"].sum()
print("La matricula de los oficiales es de", matricula_suve, "alumnos.")

matricula_suve_varones = data_suvenciones["varones"].sum()
print("Con un total de", matricula_suve_varones,"alumnos varones")

data_suvenciones
```

5. Hacer un filtro para seleccionar los establecimientos de nivel Inicial que son del distrito Quilmes.
Hacer con dos filtros, debe usarse el signo "&" y colocar cada uno entre parentesis: [(filtro1) & (filtro2)]

```python
mask_quilmes = (data["distrito"] == "Quilmes") & (data["nivel"] == "Nivel Inicial")
data_quilmenios = data.loc[mask_quilmes]
data_quilmenios
```
